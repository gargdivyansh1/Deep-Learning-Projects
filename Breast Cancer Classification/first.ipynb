```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_data = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_data)
```

```python
data_frame = pd.DataFrame(breast_cancer_data.data, columns=breast_cancer_data.feature_names)
```

```python
print(data_frame)
```

```python
## and now adding the target column into the dataframe 
data_frame['label'] = breast_cancer_data.target
```

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X, Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=2, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

## Building the neural network

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
## setting the layers of Neural Network

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
## complining the neural network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
## training the neural network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

### Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

### Accuracy of the model of test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
